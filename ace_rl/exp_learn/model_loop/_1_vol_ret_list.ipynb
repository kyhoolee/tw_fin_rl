```python
from pathlib import Path
import xarray as xr
from ace_rl.core.cube_feature import save_cube
from ace_rl.core.candle_feat_ops import make_return, make_volatility

# load cube gốc
root = Path("../../../work/processed/binance").expanduser()
cube = xr.open_zarr(
    (root / "data" / "zarr_cube_intersection").as_posix(),
    consolidated=True,
)

# -- cấu hình window --
return_lookbacks = [1, 3, 5, 10, 20, 40, 80]      # bạn đổi tùy ý
vol_windows      = [5, 10, 20, 40, 80, 160, 320]

# -- generate returns --
for lookback in return_lookbacks:
    name, ds = make_return(cube, lookback=lookback)
    save_cube(ds, root, subdir="features", name=name)
    print(f"Saved return feature {name} (lag={lookback})")

# -- generate volatility --
for window in vol_windows:
    name, ds = make_volatility(cube, window=window)
    save_cube(ds, root, subdir="features", name=name)
    print(f"Saved volatility feature {name} (window={window})")
```
