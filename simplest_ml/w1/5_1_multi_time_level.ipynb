```python
# Tail EDA for m_ret_60m across tokens

# 1) Imports (robust to different CWDs)
try:
    from features_minute import default_params, make_feature_set_id, get_hour_features
except Exception:
    import sys, os
    sys.path.append("simplest_ml/w1")
    from features_minute import default_params, make_feature_set_id, get_hour_features

try:
    from pretty_helpers import load_token_df
except Exception:
    import sys, os
    sys.path.append("simplest_ml/w1")
    from pretty_helpers import load_token_df

import os, numpy as np, pandas as pd
import matplotlib.pyplot as plt

# 2) Resolve dirs + feature_set_id
derived_dir = "simplest_ml/w1/data/derived" if os.path.isdir("simplest_ml/w1/data/derived") else "data/derived"
hourly_data_dir = "simplest_ml/w1/data" if os.path.isdir("simplest_ml/w1/data") else "data"
params = default_params()
fid = make_feature_set_id(params)
print("derived_dir:", derived_dir)
print("feature_set_id:", fid)

# 3) Config
symbols = ["BTCUSDT","BNBUSDT","SOLUSDT","ETHUSDT","AVAXUSDT","TRXUSDT","ARBUSDT","SUIUSDT"]  # b MATIC
start, end = "2023-10-30", "2025-10-30"
feature_col = "m_ret_60m"  # main tail column

# 4) Helpers
def ccdf_loglog(ax, x, label, color):
    x = np.asarray(x, dtype=float)
    x = x[np.isfinite(x)]
    x = np.abs(x)
    x = x[x > 0]
    if len(x) < 100:
        return
    x_sorted = np.sort(x)
    n = len(x_sorted)
    # CCDF: P(X > t)
    y = 1.0 - np.arange(1, n + 1) / (n + 1)
    ax.plot(np.log(x_sorted), np.log(y), lw=1.2, label=label, color=color)

def hill_tail_index(abs_x, top_frac=0.01):
    """Hill estimator for Pareto tail on |x|.
       Returns alpha (tail index); larger alpha = thinner tail."""
    x = np.asarray(abs_x, dtype=float)
    x = x[np.isfinite(x)]
    x = x[x > 0]
    if len(x) < 200:
        return np.nan
    x = np.sort(x)
    k = max(50, int(top_frac * len(x)))
    x_tail = x[-k:]
    u = x_tail[0]
    # gamma_hat = (1/k) * sum(log(x_i/u)), alpha = 1/gamma
    gamma = np.mean(np.log(x_tail / u + 1e-12))
    alpha = 1.0 / gamma if gamma > 0 else np.nan
    return alpha

def monthly_exceedance(abs_x, idx, thr):
    s = pd.Series(abs_x, index=idx)
    exc = (s > thr).astype(int)
    return exc.resample("1MS").sum()

# 5) Iterate symbols
for sym in symbols:
    print(f"\n=== {sym} ===")
    df = get_hour_features(symbol=sym, start=start, end=end, feature_set_id=fid, derived_dir=derived_dir)
    if df.empty or feature_col not in df.columns:
        print("No derived feature or missing", feature_col)
        continue

    s = df[feature_col].dropna().astype(float)
    if s.empty:
        print("Empty series after dropna.")
        continue

    # Summary quantiles
    qs = [0.5, 0.9, 0.95, 0.99, 0.995, 0.999]
    qv = s.quantile(qs)
    print("Quantiles (ret 60m):")
    print(pd.Series(qv.values, index=[f"q{int(q*1000)/10:.1f}%" for q in qs]).round(6))

    # Extreme events (top 10 up/down)
    top_up = s.sort_values(ascending=False).head(10)
    top_dn = s.sort_values(ascending=True).head(10)
    print("\nTop 10 UP events:")
    print(top_up)
    print("\nTop 10 DOWN events:")
    print(top_dn)

    # Tail threshold @ 99th percentile on |ret|
    abs_ret = s.abs()
    thr = abs_ret.quantile(0.99)
    rate = (abs_ret > thr).mean()
    print(f"\n|ret60| threshold (99%): {thr:.6f}, exceedance rate: {rate*100:.2f}%")

    # Hill tail index on |ret|
    alpha = hill_tail_index(abs_ret, top_frac=0.01)
    print(f"Hill tail index alpha  {alpha:.2f}  (higher = thinner tail)")

    # Compare with hourly 'close' returns just to sanity check alignment (optional)
    try:
        dfh = load_token_df(sym, data_dir=hourly_data_dir, interval="1h", start=pd.to_datetime(start), end=pd.to_datetime(end))
        r1h = dfh["close"].pct_change().reindex(s.index)
        corr = np.corrcoef(s.fillna(0), r1h.fillna(0))[0,1]
        print(f"Corr(m_ret_60m, 1h close ret)  {corr:.3f}")
    except Exception as e:
        print("Hourly sanity check skipped due to:", e)

    # Plots: CCDF & monthly exceedance count
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    # CCDF log-log for positive and |ret|
    ccdf_loglog(axes[0], s.dropna(), label="|ret60|", color="tab:blue")
    axes[0].set_title(f"{sym} CCDF (log-log) |ret60|")
    axes[0].set_xlabel("log(|ret60|)")
    axes[0].set_ylabel("log(1 - F)")
    axes[0].grid(alpha=0.3)
    axes[0].legend()

    # Monthly exceedances > thr
    exc_m = monthly_exceedance(abs_ret.values, s.index, thr)
    axes[1].bar(exc_m.index, exc_m.values, width=20, color="tab:gray", alpha=0.8)
    axes[1].set_title(f"{sym} monthly exceedances (|ret60| > q99)")
    axes[1].set_ylabel("count")
    axes[1].grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Systemic events: hours where many tokens simultaneously exceed their own q99(|ret60|)
import os, numpy as np, pandas as pd

# 1) Resolve derived dir + feature id (ging cell trc)
try:
    from features_minute import default_params, make_feature_set_id, get_hour_features
except Exception:
    import sys
    sys.path.append("simplest_ml/w1")
    from features_minute import default_params, make_feature_set_id, get_hour_features

derived_dir = "simplest_ml/w1/data/derived" if os.path.isdir("simplest_ml/w1/data/derived") else "data/derived"
params = default_params(); fid = make_feature_set_id(params)

symbols = ["BTCUSDT","BNBUSDT","SOLUSDT","ETHUSDT","AVAXUSDT","TRXUSDT","ARBUSDT","SUIUSDT"]
start, end = "2023-10-30", "2025-10-30"
feature_col = "m_ret_60m"

# 2) Load per-token Series and per-token q99 threshold
series = {}
thr_map = {}
for sym in symbols:
    df = get_hour_features(symbol=sym, start=start, end=end, feature_set_id=fid, derived_dir=derived_dir)
    if df.empty or feature_col not in df.columns:
        continue
    s = df[feature_col].dropna().astype(float)
    if s.empty:
        continue
    series[sym] = s
    thr_map[sym] = s.abs().quantile(0.99)

if not series:
    print("No data across symbols.")
else:
    # 3) Align into a single DataFrame with columns = tokens
    df_all = pd.concat(series, axis=1).sort_index()
    df_all = df_all.dropna(how="all")
    # Broadcast per-column thresholds
    thr_s = pd.Series(thr_map)  # index = token
    # True/False matrix where |ret| > per-token q99
    exc_mat = df_all.abs().gt(thr_s, axis=1)
    # Count exceedances per timestamp
    exc_count = exc_mat.sum(axis=1).sort_values(ascending=False)

    top_n = 10
    print(f"Top {top_n} systemic timestamps (most tokens exceed q99(|ret60|)):")
    for ts, cnt in exc_count.head(top_n).items():
        # list tokens that exceeded, with their returns
        exceeded = exc_mat.columns[exc_mat.loc[ts].values.astype(bool)]
        vals = [f"{sym}:{df_all.loc[ts, sym]:+0.4f}" for sym in exceeded]
        print(f"{ts} -> {cnt} tokens: {', '.join(vals)}")
```
